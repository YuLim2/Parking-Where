```python
# pandas 불러오기
import pandas as pd
```

```python
# 데이터 불러오기
data = pd.read_csv('./park.csv')
data.head()
```

```python
parking_map = g.Map(location=[lat, lon], 
                    zoom_start=11)
parking_map
```

```python
# 각 데이터의 위도와 경도를 map에 표시
for i in range(len(data)):
    marker = g.Marker([data.loc[i]['위도'], data.loc[i]['경도']],
                     icon = g.Icon(color='blue'),
                     popup = (data.loc[i]['주차장명'], data.loc[i]['전화번호']))
    marker.add_to(parking_map)
```

```python
parking_map
```

```python
# popup size, 내용 변경
# 여기서 오류 엄청 떳는데 앞에 내용이랑 겹쳐서 그랬음 ㅠㅠ
for i in range(len(data)):
        name = ('주차장명: ' + str(data.loc[i]['주차장명']) + '<br>'+ '전화번호: ' + str(data.loc[i]['전화번호']))
        popup = g.Popup(name, min_width=300, max_width=300)
        marker = g.Marker([data.loc[i]['위도'], data.loc[i]['경도']],
                          icon= g.Icon(color='cadetblue', icon='fa-car', prefix='fa'),
                          popup = popup).add_to(parking_map)
```

```python
parking_map
```

```python
parking_map.save("sample.html")
```

```python
# 거리 추가
for i in range(len(data)):
        place = [data.loc[i]['위도'], data.loc[i]['경도']]
        mile = round(haversine(mylocation, place, unit='km'),1)
        name = ('주차장명: ' + str(data.loc[i]['주차장명']) + '<br>'+ '전화번호: ' + str(data.loc[i]['전화번호']) + '<br>' + '주소: ' + str(data.loc[i]['주소']) + '<br>' + '거리: ' + str(mile) + 'km')
        popup = g.Popup(name, min_width=300, max_width=300)
        marker = g.Marker(place,
                          icon= g.Icon(color='cadetblue', icon='fa-car', prefix='fa'),
                          popup = popup).add_to(parking_map)
g.Marker(mylocation,
        icon = g.Icon(color='blue', icon='fa-male', prefix='fa'),
        popup = g.Popup('내 위치', min_width=40, max_width=300)).add_to(parking_map)
```
